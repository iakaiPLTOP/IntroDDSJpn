Monty Hall問題

1. TVショーの司会者Monty Hallは3つの箱を用意して、賞金をそのうちの1つの箱に隠し、残りの箱には「ハズレ」のたわしを隠した。
2. プレーヤは1つの箱を選択する。
3. 司会者は残りの箱を1つ開けて「ハズレ」であることプレーヤーに「選択する箱を変えても良い」と伝える。

ここでプレーヤは選択を変えるべきでしょうか? それとも自分の最初の選択を信じるべきでしょうか?

```python
############################################
# 用意する箱の数
num_box = 3
#
############################################
# 箱の一つに賞金をセットします。
def MontyHallSetting():
  return np.random.randint(low=1, high=num_box+1, size=(1,))[0]

############################################
# Monty Hallが箱を1つ開けますが、賞金の入った箱と、
# プレーヤが選んだ箱は開けてはいけない
def MontyHallBoxOpen(Setting, FirstBet):
  BoxNo = np.random.randint(low=1, high=num_box+1, size=(1,))[0]
  #
  # 賞金の入った箱と、プレーヤが選んだ箱は開けてはいけない
  while (BoxNo == Setting)or(BoxNo == FirstBet):
    BoxNo = np.random.randint(low=1, high=num_box+1, size=(1,))[0]
  #
  return BoxNo

############################################
# 箱の一つを選択します。
def PlayerFirstBetting():
  return np.random.randint(low=1, high=num_box+1, size=(1,))[0]

############################################
# 箱を変更します。
# ただしMonty Hallが開けた箱は選べない。変更するので、
# プレーヤが最初に選んだ箱ではない箱を選ぶ。
def PlayerSecondBetting(FirstBet, BoxNo):
  SecondBet = np.random.randint(low=1, high=num_box+1, size=(1,))[0]
  #
  # 箱を変更する。
  # Monty Hallが開けた箱は選べない。変更するので、プレーヤが最初に選んだ箱ではない箱を選ぶ。
  while (SecondBet == FirstBet)or(SecondBet == BoxNo):
    SecondBet = np.random.randint(low=1, high=num_box+1, size=(1,))[0]
  #
  return SecondBet
```

Monty Hall問題を指定回数繰り返してみます。

```python
############################################
Trials      = 1000
GetAtFirst  = 0
GetAtSecond = 0
#
for loop in range(Trials):
  #
  FlagAtFirst  = ' '
  FlagAtSecond = ' '
  #
  # Monty Hallが箱に賞金を隠します。
  Setting  = MontyHallSetting()
  #
  # Playerが箱を選びます。
  FirstBet = PlayerFirstBetting()
  #
  # 最初に選んだ箱に賞金が入っているか?
  if Setting == FirstBet:
    GetAtFirst  = GetAtFirst + 1
    FlagAtFirst = '*'
  #
  # Monty Hallが箱を開けます。
  BoxOpen = MontyHallBoxOpen(Setting, FirstBet)
  #
  # Playerが箱を選びます。
  SecondBet = PlayerSecondBetting(FirstBet, BoxOpen)
  #
  # 変えた箱に賞金が入っているか?
  if Setting == SecondBet:
    GetAtSecond  = GetAtSecond + 1
    FlagAtSecond = '*'
  #
  print( 'Setting:%d, FirstBet:%d(%c), BoxOpen:%d, SecondBet:%d(%c)' %
        (Setting, FirstBet, FlagAtFirst, BoxOpen, SecondBet, FlagAtSecond) )
```

最初の選択した場合と、選択を変えた場合の確率を計算

```python
############################################
print( "First  bet: %.2f" % (100.0 * GetAtFirst / Trials) )
print( "Second bet: %.2f" % (100.0 * GetAtSecond / Trials) )
```
